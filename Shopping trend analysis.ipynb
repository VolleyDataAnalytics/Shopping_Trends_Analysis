```python
data["Age"].plot(kind ="hist")
plt.title("Histogram of Age")
plt.show()
```

```python
data.columns
```

```python
data["Gender"].value_counts().plot(kind="bar")
plt.title("Barplot of gender distribution")
plt.show()
```

```python
data.groupby("Category")["Purchase Amount (USD)"].mean().sort_values(ascending =False).head(2).tail(3)
```

```python
data.sort_values("Review Rating", ascending =False,inplace = True)
## It WILL  SORT THE DATA IN DESCENDING  ORDER ACCORDING TO REVIEW RATING
```

```python
data
```

```python
data["Subscription Status"].value_counts().plot(kind ="pie")
plt.title("Pie chart of subscription")
plt.show()
```

```python
data["Color"].value_counts().head().plot(kind="pie")
```

```python
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(data=data, x='Gender')
plt.title('Gender Distribution')
plt.show()
```

```python
sns.boxplot(data=data, x='Gender', y='Purchase Amount (USD)')
plt.title('Purchase Amount by Gender')
plt.show()
```

```python
sns.barplot(data=data, x='Season', y='Purchase Amount (USD)', ci=None)
plt.title('Average Purchase Amount by Season')
plt.show()
```

```python
sns.pairplot(data=data[['Age', 'Purchase Amount (USD)', 'Previous Purchases', 'Review Rating']])
plt.suptitle('Pairplot of Numeric Variables')
plt.show()
```

```python
payment_method_counts = data['Payment Method'].value_counts()
plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.2f%%', startangle=140)
plt.axis('equal')
plt.title('Payment Method Distribution')
plt.show()
```

```python
sns.boxplot(data=data, x='Subscription Status', y='Purchase Amount (USD)')
plt.title('Purchase Amount by Subscription Status')
plt.show()
```

```python
sns.countplot(data=data, x='Discount Applied')
plt.title('Count of Discount Applied')
plt.show()
```

```python
sns.barplot(data=data, x='Category', y='Review Rating')
plt.title('Average Review Rating by Category')
plt.show()
```

```python
sns.violinplot(data=data, x='Frequency of Purchases', y='Purchase Amount (USD)')
plt.title('Purchase Amount by Frequency of Purchases')
plt.xticks(rotation=90)
plt.show()
```
